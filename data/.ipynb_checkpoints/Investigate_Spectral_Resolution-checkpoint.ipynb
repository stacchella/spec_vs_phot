Proof that we need R~2700 instead of R~1000.
Assume absorption and emission line are both Gaussian, no velocity offset, redshift known. 
Furthermore, sigma_a constant and larger than sigma_e.
Variation in A_a, sigma_a, A_e, and sigma_e.

sigma_a = 200-400 km/s => 2.67-5.33 A at 4000 A
sigma_e = 20-140 km/s  => 0.27-1.87 A at 4000 A

```python
# import modules

import numpy as np
import scipy.optimize as opt
from astropy.convolution import Gaussian1DKernel, convolve

import matplotlib.pyplot as plt
plt.rcParams['mathtext.fontset'] = 'stixsans'
plt.rcParams['font.family'] = 'sans'
plt.rcParams['font.serif'] = 'STIXGeneral'
```

```python
# define paths

path_figure = '/Users/sandrotacchella/ASTRO/JWST/Proposal_GO_Cycle1/figures/'
```

```python
# set parameters

wavelength_line = 4102.89  # look at Hdelta
redshift = 2.241

wave_min, wave_max = 4060.0, 4140.0
contiuum_level = 1.0

# compute smoothing kernel
dlam = 0.23
R_h = 2700.0
R_l = 1000.0
dlam_res_h = (1+redshift)*wavelength_line/R_h
dlam_res_l = (1+redshift)*wavelength_line/R_l
dpix_res_h = dlam_res_h/dlam
dpix_res_l = dlam_res_l/dlam
g_h = Gaussian1DKernel(stddev=dpix_res_h)
g_l = Gaussian1DKernel(stddev=dpix_res_l)

# SNR
snr = 2.0  # assume constant for the moment, in the continuum
```

```python
def gaussian(x, mu, sig):
    '''
    normalized Gaussian
    '''
    return 1.0/np.sqrt(2*np.pi*np.power(sig, 2.))*np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))



def set_line(x, mu, A_a, sig_a, A_e, sig_e):
    return contiuum_level-A_a*gaussian(x, mu, sig_a)+A_e*gaussian(x, mu, sig_e)



def fitting_fct(x, A_a, sig_a, A_e, sig_e):
    return contiuum_level-A_a*gaussian(x, wavelength_line*(1.0+redshift), sig_a)+A_e*gaussian(x, wavelength_line*(1.0+redshift), sig_e)

```

```python
def calculate_EW(wave, line, A, sig):
    return(np.trapz(1.0-A*gaussian(wave, line, sig)/contiuum_level, wave))


def get_EW_with_uncertainty(wave, A_a, A_a_unc, sig_a, sig_a_unc, num_sample=100):
    num_sample = 100
    A_a_fit = np.random.normal(loc=A_a, scale=A_a_unc, size=num_sample)
    sig_a_fit = np.random.normal(loc=sig_a, scale=sig_a_unc, size=num_sample)
    EW_list = []
    for ii in range(num_sample):
        EW_list.append(calculate_EW(wave, wavelength_line*(1.0+redshift), A_a_fit[ii], sig_a_fit[ii])/(1.0+redshift))
    EW = np.percentile(EW_list, 50)
    EW_p16 = np.percentile(EW_list, 16)
    EW_p84 = np.percentile(EW_list, 84)
    return(EW, EW_p16, EW_p84)


def run_simulation(A_a, sig_a, A_e, sig_e, plot_fit=False, save_fig=False):
    print '========================'
    print 'input values: '
    print A_a, sig_a, A_e, sig_e
    print '------------------------'
    # set line and compute EW
    wave = np.linspace(wave_min, wave_max, num=10000)
    line = set_line(wave, wavelength_line, A_a, sig_a, A_e, sig_e)
    EW_line_abs = calculate_EW(wave, wavelength_line, A_a, sig_a)
    print 'true EW = ', round(EW_line_abs, 1)
    # redshift line
    wave_z = wave*(1.0+redshift)
    # grid according to JWST NIRSpec IFU: 
    wave_IFU = np.arange(np.min(wave_z), np.max(wave_z), dlam)
    line_IFU = np.interp(wave_IFU, wave_z, line)
    # smooth according to resolution R
    line_IFU_R_h = convolve(line_IFU, g_h, boundary='extend')
    line_IFU_R_l = convolve(line_IFU, g_l, boundary='extend')
    # add noise
    snoise_sigma_h = contiuum_level / snr
    snoise_sigma_l = contiuum_level / snr
    snoise_h = np.random.normal(0, 1.0, size=len(line_IFU_R_h)) * snoise_sigma_h
    snoise_l = np.random.normal(0, 1.0, size=len(line_IFU_R_l)) * snoise_sigma_l
    line_IFU_wn_R_h = line_IFU_R_h + snoise_h
    line_IFU_wn_R_l = line_IFU_R_l + snoise_l
    # fit data
    upper_bound = [100.0, 20.0, 100.0, 20.0]
    optimizedParameters_wn_h, pcov_wn_h = opt.curve_fit(fitting_fct, wave_IFU, line_IFU_wn_R_h, p0=[A_a, sig_a, A_e, sig_e], bounds=([0.0, 0.0, 0.0, 0.0], upper_bound))
    optimizedParameters_wn_l, pcov_wn_l = opt.curve_fit(fitting_fct, wave_IFU, line_IFU_wn_R_l, p0=[A_a, sig_a, A_e, sig_e], bounds=([0.0, 0.0, 0.0, 0.0], upper_bound))
    print optimizedParameters_wn_h
    #print np.sqrt(pcov_wn_h)
    print optimizedParameters_wn_l
    #print np.sqrt(pcov_wn_l)
    # check if valid fit
    hit_bound_h = []
    hit_bound_l = []
    for ii in range(len(optimizedParameters_wn_h)):
        hit_bound_h.append(np.sum(optimizedParameters_wn_h[ii] > 0.95 * upper_bound[ii]))
        hit_bound_l.append(np.sum(optimizedParameters_wn_l[ii] > 0.95 * upper_bound[ii]))
    bad_fit_h = (np.sum(hit_bound_h)>0.0)
    bad_fit_l = (np.sum(hit_bound_l)>0.0)
    # estimate EW including uncertainty
    EW_line_abs_h, EW_line_abs_h_p16, EW_line_abs_h_p84 = get_EW_with_uncertainty(wave_IFU, optimizedParameters_wn_h[0], np.sqrt(pcov_wn_h[0][0]), optimizedParameters_wn_h[1], np.sqrt(pcov_wn_h[1][1]), num_sample=100)
    EW_line_abs_l, EW_line_abs_l_p16, EW_line_abs_l_p84 = get_EW_with_uncertainty(wave_IFU, optimizedParameters_wn_l[0], np.sqrt(pcov_wn_l[0][0]), optimizedParameters_wn_l[1], np.sqrt(pcov_wn_l[1][1]), num_sample=100)
    print 'high res EW = ', round(EW_line_abs_h, 1)
    print 'low res EW = ', round(EW_line_abs_l, 1)
    if plot_fit:
        f, (ax1, ax2) = plt.subplots(2, 1, figsize=(7.0, 5.0))
        ax1.plot(wave_IFU, line_IFU, '--', lw=2, color='black', label='input', zorder=5)
        ax1.plot(wave_IFU, line_IFU_wn_R_h, color='C3', alpha=0.8, label='R=2700 (with noise)')
        ax1.plot(wave_IFU, line_IFU_R_h, '-', color='white', lw=8, alpha=0.7)
        ax1.plot(wave_IFU, line_IFU_R_h, '-', color='C3', lw=2, alpha=1.0, label='R=2700')
        #ax1.plot(wave_IFU, fitting_fct(wave_IFU, *optimizedParameters_wn_h), '--', color='C3')
        ax2.plot(wave_IFU, line_IFU, '--', lw=2, color='black', label='input', zorder=5)
        ax2.plot(wave_IFU, line_IFU_wn_R_l, color='C0', alpha=0.8, label='R=1000 (with noise)')
        ax2.plot(wave_IFU, line_IFU_R_l, '-', color='white', lw=8, alpha=0.7)
        ax2.plot(wave_IFU, line_IFU_R_l, '-', color='C0', lw=2, alpha=1.0, label='R=1000')
        #axes.plot(wave_IFU, fitting_fct(wave_IFU, *optimizedParameters_wn_l), '--', color='C0')
        plt.subplots_adjust(hspace=0.05)
        ax1.set_xlim([13225, 13375])
        ax2.set_xlim([13225, 13375])
        ax1.set_ylim([-1.2, 2.2])
        ax2.set_ylim([-1.2, 2.2])
        ax1.set_ylabel(r'$\mathrm{flux}\/\/\/[\mathrm{arb.\/units}]$', fontsize=18)
        ax2.set_xlabel(r'$\lambda_{\rm obs}\/\/\/[\mathrm{\AA}]$', fontsize=18)
        ax2.set_ylabel(r'$\mathrm{flux}\/\/\/[\mathrm{arb.\/units}]$', fontsize=18)
        ax1.xaxis.set_ticklabels([]) # Hide only x axis
        if save_fig:
            plt.savefig(path_figure + 'investigate_resolution_spec.pdf', format='pdf', bbox_inches='tight')
        plt.show()
    return(EW_line_abs, [EW_line_abs_h, EW_line_abs_h_p16, EW_line_abs_h_p84], [EW_line_abs_l, EW_line_abs_l_p16, EW_line_abs_l_p84], bad_fit_h, bad_fit_l)
```

```python
# example

run_simulation(25, 4.3239308946120065, 10, 1.4507115558465888, plot_fit=True, save_fig=True)
```

```python

# set up grid
num_draws = 10000
A_a_list = np.random.uniform(low=15, high=30, size=num_draws)
sig_a_list = np.random.uniform(low=2.67, high=5.33, size=num_draws)
A_e_list = np.random.uniform(low=5, high=15, size=num_draws)
sig_e_list = np.random.uniform(low=0.27, high=1.87, size=num_draws)


EW_input, EW_output_h, EW_output_h_p16, EW_output_h_p84, EW_output_l, EW_output_l_p16, EW_output_l_p84, bad_fit_h, bad_fit_l = [], [], [], [], [], [], [], [], []

for ii in range(num_draws):
    output = run_simulation(A_a_list[ii], sig_a_list[ii], A_e_list[ii], sig_e_list[ii], plot_fit=False)
    EW_input = np.append(EW_input, output[0])
    EW_output_h = np.append(EW_output_h, output[1][0])
    EW_output_h_p16 = np.append(EW_output_h_p16, output[1][1])
    EW_output_h_p84 = np.append(EW_output_h_p84, output[1][2])
    EW_output_l = np.append(EW_output_l, output[2][0])
    EW_output_l_p16 = np.append(EW_output_l_p16, output[2][1])
    EW_output_l_p84 = np.append(EW_output_l_p84, output[2][2])
    bad_fit_h = np.append(bad_fit_h, output[3])
    bad_fit_l = np.append(bad_fit_l, output[4])

```

```python
# mask bad fits

good_fits_h = (bad_fit_h==0.0)
good_fits_l = (bad_fit_l==0.0)

print 'number of bad fits (high res): ', np.sum(bad_fit_h>0.0)
print 'fraction of bad fits (high res): ', 1.0*np.sum(bad_fit_h>0.0)/len(bad_fit_h)
print 'number of bad fits (low res): ', np.sum(bad_fit_l>0.0)
print 'fraction of bad fits (low res): ', 1.0*np.sum(bad_fit_l>0.0)/len(bad_fit_l)

EW_in_h, EW_out_h, EW_out_h_p16, EW_out_h_p84 = EW_input[good_fits_h], EW_output_h[good_fits_h], EW_output_h_p16[good_fits_h], EW_output_h_p84[good_fits_h]      
EW_in_l, EW_out_l, EW_out_l_p16, EW_out_l_p84 = EW_input[good_fits_l], EW_output_l[good_fits_l], EW_output_l_p16[good_fits_l], EW_output_l_p84[good_fits_l]

```

```python
# investigate fit uncertainty

err_h_up, err_h_down = EW_out_h_p84-EW_out_h, EW_out_h-EW_out_h_p16
err_l_up, err_l_down = EW_out_l_p84-EW_out_l, EW_out_l-EW_out_l_p16
```

```python
# Figure input vs output

f, axes = plt.subplots(1, 1, figsize=(5.0, 5.0))

axes.plot([0.0, 100.0], [0.0, 100.0], '-', lw=3, color='black')
axes.errorbar(EW_in_h, EW_out_h, fmt='.', color='C3', lw=2, alpha=0.5, label='R=2700')  # yerr=[err_h_up, err_h_down], 
axes.errorbar(EW_in_l, EW_out_l, fmt='.', color='C0', lw=2, alpha=0.5, label='R=1000')  # yerr=[err_l_up, err_l_down], 

axes.errorbar(68, 55, yerr=[[np.median(err_h_up)], [np.median(err_h_down)]], fmt='o', color='C3')
axes.errorbar(66, 55, yerr=[[np.median(err_l_up)], [np.median(err_l_down)]], fmt='o', color='C0')


axes.set_xlim([45, 70])
axes.set_ylim([45, 70])
axes.set_xlabel(r'$\mathrm{EW_{\rm input}}\/\/\/[\mathrm{\AA}]$', fontsize=18)
axes.set_ylabel(r'$\mathrm{EW_{\rm recovered}}\/\/\/[\mathrm{\AA}]$', fontsize=18)
plt.legend(frameon=False, fontsize=14)
plt.savefig(path_figure + 'investigate_resolution_EW.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
